```python
# Creating function
def plot_line(W, w0, x,y):
    min = x[:,0].min()
    max = x[:,0].max()

    w1 = W[0]
    w2 = W[1]
    w0 = w0

    slope = -w2/w1
    intercept = -w0/w1

    x_values = np.linspace(min, max, 100)
    y_values = slope * x_values + intercept

    sns.scatterplot(x = x[:,0], y = x[:,1], hue = y)
    plt.plot(x_values, y_values, c = 'r')
    plt.show()
```

```python
def distance_from_line(W,w0,xi):
    distance = (W@xi + w0)/np.linalg.norm(W)
    return distance
```

```python
W = np.array([1,1])
w0 = 1
distance_from_line(W, w0, [15,1])
```

```python
distance_from_line(W,w0,[6,-8])
```

```python
y[y==0] = -1
```

```python
y
```

```python
def perceptron(W, w0, x,y, iter=10):
    for i in range(iter):
        misclassification = 0
        for xi,yi in zip(x,y):
            d = distance_from_line(W,w0,xi)
            if np.sign(d) != yi:
                misclassification += 1
                W = W + (xi * yi)
                w0 = w0 + yi # why
            plot_line(W,w0,x,y)
            if misclassification == 0:
                return W,w0

    return W,w0
                
    
```

```python
perceptron(W,w0,x,y,iter=10)
```
