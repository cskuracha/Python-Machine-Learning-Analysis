## **Gaussian Distribution**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math as m
from scipy.stats import norm
```

```python
np.random.randint(0,2)
```

```python
coin_toss = np.random.randint(0,2,size = (2, 10))
coin_toss
```

```python
head_count = coin_toss.sum(axis = 1)
head_count
```

```python
sns.countplot(head_count)
```

```python
coin_toss = np.random.randint(0,2,size = (30000,10))
coin_toss
```

```python
head_count = coin_toss.sum(axis = 1)
head_count
```

```python
head_count.shape
```

```python
sns.countplot(x=head_count)
```

#### **Rolling a dice, twice**

```python
dice_roll = np.random.randint(1,7, size=(30000,2))
dice_roll
```

```python
dice_count = dice_roll.sum(axis = 1)
dice_count
```

```python
sns.countplot(x=dice_count)
```

```python
dice_count1 = np.random.randint(1,7, size = (30000,1))
dice_count1
```

```python
dice_count = dice_count1.sum(axis = 1)
```

```python
sns.countplot(x = dice_count)
```
