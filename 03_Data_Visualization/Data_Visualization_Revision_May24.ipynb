```python
ice_hockey = data.loc[data['Name']=='Ice Hockey']
```

```python
sns.lineplot(x='Year', y = 'NA_Sales', data = ice_hockey)
```

```python
sns.lineplot(x='Year', y = 'EU_Sales', data = ice_hockey)
```

```python
sns.lineplot(x="Year", y = "NA_Sales", data = ice_hockey, color = "black")
plt.xlabel("Year")
plt.ylabel("NA_Sales")
plt.show()
```

```python
sns.lineplot(x="Year",y="NA_Sales", data = ice_hockey, color = "black")
plt.xlim(left=2000,right= 2015)
plt.xlabel("Years")
plt.ylabel("NA_Sales")
plt.show()
```

```python
baseball = data.loc[data["Name"]== "Baseball"]
baseball.head()
```

```python
sns.lineplot(x="Year",y="NA_Sales",data = baseball, color = "Red")
plt.xlabel("Year")
plt.ylabel("NA_Sales")
plt.show()
```

```python
sns.lineplot(x="Year",y="NA_Sales", data = ice_hockey, color = "black", label="Ice Hockey")
sns.lineplot(x="Year",y="NA_Sales", data = baseball, color = "red", label = "Baseball")
plt.legend()
plt.show()
```

```python
sns.lineplot(x="Year",y="NA_Sales", data = ice_hockey, color = "black", label="Ice Hockey")
sns.lineplot(x="Year",y="NA_Sales", data = baseball, color = "red", label = "Baseball")
plt.legend(loc = "lower left")
plt.show()
```

```python
sns.lineplot(x="Year",y="NA_Sales", data = ice_hockey, color = "black", label="Ice Hockey")
sns.lineplot(x="Year",y="NA_Sales", data = baseball, color = "red", label = "Baseball")
plt.legend(loc = "lower right")
plt.show()
```

```python
sns.lineplot(x="Year",y="NA_Sales", data = ice_hockey, color = "black", label="Ice Hockey")
sns.lineplot(x="Year",y="NA_Sales", data = baseball, color = "red", label = "Baseball")
plt.legend(loc = (-.5,-0.5))
plt.show()
```

```python
sns.lineplot(x="Year",y="NA_Sales", data = ice_hockey, color = "black", label="Ice Hockey")
sns.lineplot(x="Year",y="NA_Sales", data = baseball, color = "red", label = "Baseball")
plt.legend(loc = (-.5,-0.5), ncol = 2, frameon = False)
plt.show()
```

```python
sns.lineplot(x="Year",y="NA_Sales", data = ice_hockey, color = "black", label="Ice Hockey")
sns.lineplot(x="Year",y="NA_Sales", data = baseball, color = "red", label = "Baseball")
plt.legend(loc = (-.5,-0.5), ncol = 2, frameon = False)
plt.text(2015, max(ice_hockey["NA_Sales"]), "Max Sales")
plt.show()
```
