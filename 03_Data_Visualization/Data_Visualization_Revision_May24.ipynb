### **Top 3 Publisher**

```python
data.head()
```

```python
top3_pub = data["Publisher"].value_counts()[:3].index
top3_pub
```

```python
top3_gen = data["Genre"].value_counts()[:3].index
top3_gen
```

```python
top3_platform = data["Platform"].value_counts()[:3].index
top3_platform
```

```python
data.loc[data["Publisher"].isin(top3_pub)]
```

```python
data.loc[data['Genre'].isin(top3_gen)]
```

```python
data.loc[data['Platform'].isin(top3_platform)]
```

```python
top3_data = data.loc[data['Publisher'].isin(top3_pub) & data['Genre'].isin(top3_gen) & data['Platform'].isin(top3_platform)]
top3_data
```
