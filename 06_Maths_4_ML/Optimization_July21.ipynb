```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_partial(func,x,y):
  delta =0.0000001
  fx = (func(x+delta,y) - func(x,y))/delta
  fy = (func(x,y+delta) - func(x,y))/delta
  return fx,fy
```

```python
def func(x,y):
  return x**2+y**2
```

```python
get_partial(func,1,2)
```
