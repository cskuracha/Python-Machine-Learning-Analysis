```python
# Creating function
def plot_line(W, w0, x,y):
    min = x[:,0].min()
    max = x[:,0].max()

    w1 = W[0]
    w2 = W[1]
    w0 = w0

    slope = -w2/w1
    intercept = -w0/w1

    x_values = np.linspace(min, max, 100)
    y_values = slope * x_values + intercept

    sns.scatterplot(x = x[:,0], y = x[:,1], hue = y)
    plt.plot(x_values, y_values, c = 'r')
    plt.show()
```
