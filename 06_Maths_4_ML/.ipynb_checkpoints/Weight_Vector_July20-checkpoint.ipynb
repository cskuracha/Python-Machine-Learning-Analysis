```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Generate a 2D linearly separable dataset
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, 
                           n_informative=2, n_clusters_per_class=1, 
                           random_state=42)

# Train a linear Support Vector Machine
model = SVC(kernel='linear', C=1.0)
model.fit(X, y)

# Extract weight vector and bias
w = model.coef_[0]  # Weight vector [w1, w2]
b = model.intercept_[0]  # Bias term

print(f"Weight Vector: {w}")
print(f"Bias: {b}")
```

```python
# Method 1: Using Two Points on the Decision Boundary
# Create the decision boundary line: w1*x + w2*y + b = 0
# Solve for y: y = (-w1*x - b) / w2

x_range = np.linspace(X[:, 0].min()-1, X[:, 0].max()+1, 100)
y_boundary = (-w[0] * x_range - b) / w[1]

# Take two points on the boundary
point1 = np.array([x_range[20], y_boundary[20]])
point2 = np.array([x_range[80], y_boundary[80]])

# Create a vector along the boundary
boundary_vector = point2 - point1

# Check perpendicularity using dot product
dot_product = np.dot(w, boundary_vector)

print(f"\nBoundary Vector: {boundary_vector}")
print(f"Dot Product (w · boundary_vector): {dot_product}")
print(f"Is perpendicular? {np.isclose(dot_product, 0, atol=1e-10)}")
```

```python
# Method 2: Using the Direction Vector of the Line
# The decision boundary has direction vector perpendicular to w
# If w = [a, b], then direction vector of line is [-b, a] or [b, -a]

line_direction = np.array([-w[1], w[0]])  # Perpendicular to weight vector

# Verify this is perpendicular
dot_product_method2 = np.dot(w, line_direction)

print(f"\nLine Direction Vector: {line_direction}")
print(f"Dot Product (method 2): {dot_product_method2}")
print(f"Is perpendicular? {np.isclose(dot_product_method2, 0)}")
```

```python
#Method 3: Testing with Multiple Points on the Boundary
def verify_perpendicularity_multiple_points(w, b, n_points=5):
    """Test perpendicularity using multiple points on the decision boundary"""
    
    # Generate multiple x values
    x_vals = np.linspace(-3, 3, n_points)
    
    # Calculate corresponding y values on the boundary
    y_vals = (-w[0] * x_vals - b) / w[1]
    
    # Create vectors from first point to all other points
    first_point = np.array([x_vals[0], y_vals[0]])
    
    dot_products = []
    for i in range(1, n_points):
        current_point = np.array([x_vals[i], y_vals[i]])
        boundary_vector = current_point - first_point
        dot_prod = np.dot(w, boundary_vector)
        dot_products.append(dot_prod)
    
    return dot_products

# Test with multiple points
dot_products = verify_perpendicularity_multiple_points(w, b)
print(f"\nDot products with multiple boundary vectors: {dot_products}")
print(f"All close to zero? {all(np.isclose(dp, 0, atol=1e-10) for dp in dot_products)}")
```

```python
# Create visualization
plt.figure(figsize=(12, 8))

# Plot data points
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', 
                     s=50, alpha=0.7, edgecolors='black')

# Plot decision boundary
x_line = np.linspace(X[:, 0].min()-1, X[:, 0].max()+1, 100)
y_line = (-w[0] * x_line - b) / w[1]
plt.plot(x_line, y_line, 'r-', linewidth=2, label='Decision Boundary')

# Plot weight vector (scaled for visibility)
center_x, center_y = np.mean(X, axis=0)
scale_factor = 2.0  # Adjust for better visualization
plt.arrow(center_x, center_y, w[0]*scale_factor, w[1]*scale_factor,
          head_width=0.2, head_length=0.2, fc='red', ec='red',
          label='Weight Vector (scaled)')

# Plot a boundary vector for comparison
if len(x_line) > 1:
    boundary_start = np.array([x_line[30], y_line[30]])
    boundary_end = np.array([x_line[70], y_line[70]])
    boundary_vec = boundary_end - boundary_start
    plt.arrow(boundary_start[0], boundary_start[1], 
              boundary_vec[0], boundary_vec[1],
              head_width=0.15, head_length=0.15, fc='blue', ec='blue',
              label='Boundary Direction Vector')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Weight Vector Perpendicular to Decision Boundary')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.show()
```
