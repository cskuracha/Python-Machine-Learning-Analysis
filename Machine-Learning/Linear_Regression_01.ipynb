# Linear Regression
### Learn to implement the model function f for linear regression with one variable

#### Importing dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("classic")
```

```python
#plt.style.available
```

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

#### Number of training examples

```python
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is {m}")
```

#### Training Examples

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x_0, y_0) = ({x_i},{y_i})")
i = 1
x_i = x_train[i]
y_i = y_train[i]
print(f"(x_1, y_1) = ({x_i},{y_i})")
```

#### Plotting the data

```python
plt.scatter(x_train, y_train, marker = "x", c = "r")
plt.title("House Pricing")
plt.ylabel("Price ( in 1000 dollar)")
plt.xlabel("Size (in sqft)")
plt.show()
```

#### Model Function

```python
(w,b) = (100,100)
print(f"w: {w}")
print(f"b: {b}")
```

#### Compute models

```python
def compute_model_output(x,w,b):
    """Compute the prediction of a linear model"""
    m = x.shape[0]
    print(f"Value of m is {m}")
    f_wb = np.zeros(m)
    print(f"Initial value of f_wb is {f_wb}")
    for i in range(m):
        f_wb[i] = w * x[i] + b
        print(f"Value of f_wb for {i} is {f_wb[i]}")
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)
print(f"Value of tmp_f_wb: {tmp_f_wb}")
plt.plot(x_train, tmp_f_wb, c = 'b', label = "Our Prediction")
plt.scatter(x_train, y_train, marker = "x", c = "r", label = "Actual Values")
plt.title("Housing Prices")
plt.ylabel("Price")
plt.xlabel("Size")
plt.legend()
plt.show()
```

```python
tmp_f_wb = compute_model_output(x_train, 200, 200)
print(f"Value of tmp_f_wb : {tmp_f_wb}")
plt.plot(x_train, tmp_f_wb, c = 'b', label = "Our Prediction")
plt.scatter(x_train, y_train, marker = "x", c = "r", label = "Actual Values")
plt.title("Housing Prices")
plt.xlabel("Size")
plt.ylabel("Price")
plt.legend()
plt.show()
```
