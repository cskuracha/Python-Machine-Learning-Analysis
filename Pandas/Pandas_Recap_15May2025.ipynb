#### **How to find the movies by high budget directors**

```python
data.groupby("director_name")['budget'].max() # Return series
```

```python
data.groupby("director_name")['budget'].max().reset_index()
```

```python
data_dir_budget = data.groupby('director_name')['budget'].max().reset_index()
data_high_budget = data_dir_budget.loc[data_dir_budget['budget']>=100000000]
data_high_budget
```

```python
data.loc[data['director_name'].isin(data_high_budget['director_name'])]
```

#### **Above cells can be combined using**

```python
data.groupby('director_name').filter(lambda x: x['budget'].max()>=100000000)
```

#### **Identify movies by Risky directors** 

```python
def func(x):
    x['risky'] = x['budget']-x['revenue'].mean() >=0
    return x
```

```python
data_risky = data.groupby('director_name',group_keys = False).apply(func)
data_risky
```

```python
data_risky.loc[data_risky['risky'] == True]
```
